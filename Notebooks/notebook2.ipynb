## Ellipse & Circles

```python
# Ellipse 1: Center at (2, 3) and a=4, b=3, gives (x-2)**2/4**2 + (y-3)**2/3**2 = 1
ellipse1 = "(x - 2)**2 / 16 + (y - 3)**2 / 9 - 1"
# Ellipse 2: Center at (-1, -1) and a=3, b=2, gives (x+1)**2/3**2 + (y+1)**2/2**2 = 1
ellipse2 = "(x + 1)**2 / 9 + (y + 1)**2 / 4 - 1"
```

```python
e1 = Conic.create(ellipse1)
print(f"{e1.__repr__()}\ntype: {e1.type}\nInstance: {type(e1)}"
      f"\nCoefficients: {e1.coeff}\nEquation: {e1}")
print(f"Matrix of Coefficients:")
e1.print_matrix()
e1.plot()
```

```python
e2 = Conic.create(ellipse2)
print(f"{e2.__repr__()}\ntype: {e2.type}\nInstance: {type(e2)}\nCoefficients: {e2.coeff}\nEquation: {e2}")
print(f"Matrix of Coefficients:")
e2.print_matrix()
e2.plot()
```

Some 'circles' below are actually not!

```python
# Circle 1: General form with fractional coefficients, equivalent to a circle centered at (3, -2) with radius sqrt(35)
circle1 = "x**2/9 - 2*x/3 + y**2/4 + y/2 - 35/36"

# Circle 2: General form with fractional coefficients, equivalent to a circle centered at (-5, 3) with radius sqrt(14)
circle2 = "x**2/16 + 5*x/8 + y**2/9 - y/3 - 14/64"

# Circle 3: General form with fractional coefficients, equivalent to a circle centered at (1, -4) with radius sqrt(20)
circle3 = "x**2/25 - x/25 + y**2/16 + 2*y/4 - 20/625"

# Circle 4: Center at (2, 3) and radius 4, gives (x-2)**2 + (y-3)**2 = 4**2
circle4 = "(x - 2)**2 + (y - 3)**2 - 16"

# Circle 5: Center at (-1, -1) and radius 3, gives (x+1)**2 + (y+1)**2 = 3**2
circle5 = "(x + 1)**2 + (y + 1)**2 - 9"

# Circle 6: Center at (3, 4) and radius 5, gives (x-3)**2 + (y-4)**2 = 5**2
circle6 = "(x - 3)**2 + (y - 4)**2 - 25"
```

```python
c1 = Conic.create(circle1)
c2 = Conic.create(circle2)
c3 = Conic.create(circle3)
c4 = Conic.create(circle4)
c5 = Conic.create(circle5)
c6 = Conic.create(circle6)
```

```python
circles = [c1, c2, c3, c4, c5, c6]
for i, circle in enumerate(circles, start=1):
    print("---")
    print(f"{circle.__repr__()}\ntype: {circle.type}\nInstance: {type(circle)}"
          f"\nCoefficients: {circle.coeff}\nEquation: {circle}")
    print(f"Matrix of Coefficients:")
    circle.print_matrix()
    circle.display()
    circle.plot()
```

## Parabolas

```python
# Parabola 1: General form equivalent to a parabola with vertex at (3, -2) and opens upwards
parabola1 = "y - ((x - 3)**2) / 4 + 2"

# Parabola 2: General form equivalent to a parabola with vertex at (-5, 3) and opens downwards
parabola2 = "y + ((x + 5)**2) / 3 - 3"

# Parabola 3: General form equivalent to a parabola with vertex at (1, -4) and opens upwards
parabola3 = "y - ((x - 1)**2) / 5 + 4"

# Parabola 4: Vertex at (2, 3) and opens rightwards, gives x - 2 = (y - 3)**2
parabola4 = "x - ((y - 3)**2) + 2"

# Parabola 5: Vertex at (-1, -1) and opens leftwards, gives x + 1 = (y + 1)**2
parabola5 = "x + ((y + 1)**2) - 1"

# Parabola 6: Vertex at (3, 4) and opens downwards, gives y - 4 = -(x - 3)**2
parabola6 = "y - (-(x - 3)**2) + 4"
```

```python
p1 = Conic.create(parabola1)
p2 = Conic.create(parabola2)
p3 = Conic.create(parabola3)
p4 = Conic.create(parabola4)
p5 = Conic.create(parabola5)
p6 = Conic.create(parabola6)
```

```python
parabolas = [p1, p2, p3, p4, p5, p6]
for i, parabola in enumerate(parabolas, start=1):
    print("---")
    print(f"{parabola.__repr__()}\ntype: {parabola.type}\nInstance: {type(parabola)}"
          f"\nCoefficients: {parabola.coeff}\nEquation: {parabola}")
    print(f"Matrix of Coefficients:")
    parabola.print_matrix()
    parabola.display()
    parabola.plot()
```

## Hyperbolas

```python
# Hyperbola 1: General form equivalent to a hyperbola centered at (3, -2)
hyperbola1 = "x**2/9 - y**2/4 - 2*x/3 + y/2 - 1"

# Hyperbola 2: General form equivalent to a hyperbola centered at (-5, 3)
hyperbola2 = "x**2/16 - y**2/9 + 5*x/8 - y/3 + 1"

# Hyperbola 3: General form equivalent to a hyperbola centered at (1, -4)
hyperbola3 = "x**2/25 - y**2/16 + x/25 - 2*y/4 + 1"

# Hyperbola 4: Center at (2, 3), gives (x-2)**2/16 - (y-3)**2/9 = 1
hyperbola4 = "(x - 2)**2/16 - (y - 3)**2/9 - 1"

# Hyperbola 5: Center at (-1, -1), gives (x+1)**2/9 - (y+1)**2/4 = 1
hyperbola5 = "(x + 1)**2/9 - (y + 1)**2/4 - 1"

# Hyperbola 6: Center at (3, 4), gives (x-3)**2/16 - (y-4)**2/9 = 1
hyperbola6 = "(x - 3)**2/16 - (y - 4)**2/9 - 1"
```

```python
h1 = Conic.create(hyperbola1)
h2 = Conic.create(hyperbola2)
h3 = Conic.create(hyperbola3)
h4 = Conic.create(hyperbola4)
h5 = Conic.create(hyperbola5)
h6 = Conic.create(hyperbola6)
```

```python
hyperbolas = [h1, h2, h3, h4, h5, h6]
for i, hyperbola in enumerate(hyperbolas, start=1):
    print("---")
    print(f"{hyperbola.__repr__()}\ntype: {hyperbola.type}\nInstance: {type(hyperbola)}"
          f"\nCoefficients: {hyperbola.coeff}\nEquation: {hyperbola}")
    print(f"Matrix of Coefficients:")
    hyperbola.print_matrix()
    hyperbola.display()
    hyperbola.plot()
```
