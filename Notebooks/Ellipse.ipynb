```python
import sympy
from conics.conic_factory import *
```

## 'Standard' Ellipse

```python
ellipse1 = "x**2/4 - 2*x/3 + y**2/9 + y/2 - 35/36"

e1 = Conic.create(ellipse1)
e1.get_info()
```

```python
# Translation to standard form needs to occur before certain attributes can be computed
# DEF - 'Standard Form', centre = (0, 0), orientation = 'horizontal'

e1.semimajor_axis
```

```python
e1.plot_standard()
```

#### Translation and rotation are required based on `get_info()` plot

```python
e1.translate(display=True)
```

```python
e1.rotate(display=True)
```

```python
# What did we just do?
e1.print_history()
```

```python
e1.plot_standard()
```

## Ellipse #2 - Rotated

```python
rote2 = "3*x^2 - 8*x*y + 9*y^2 - 7*x - 3*y - 9"
re2 = Conic.create(rote2)
re2.get_info()
```

```python
re2.translate(display=True)
```

```python
re2.rotate(display=True)
```

```python
re2.print_history()
```

```python
re2.plot_standard(x_range=(-7.5, 7.5), y_range=(-5, 5))
```
