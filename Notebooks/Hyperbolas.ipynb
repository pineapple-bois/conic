# Hyperbolas

```python
import sympy
from conics.conic_factory import *
```

----
&nbsp;
#### Example 1

```python
# Hyperbola 1: opens horizontally (x term is positive)
hyperbola1 = "x**2/9 - y**2/4 - 2*x/3 + y/2 - 1"
h1 = Conic.create(hyperbola1)
h1.get_info()
```

```python
h1.translate(display=True)
```

```python
h1.print_history()
```

```python
h1.plot_standard()
```

----
&nbsp;
#### Example 2

```python
# Hyperbola 2: opens vertically (y term is positive)
hyperbola2 = "(y-1)**2/16 - (x+1)**2/9 - 1"
h2 = Conic.create(hyperbola2)
h2.get_info()
```

```python
h2.rotate(display=True)
h2.translate(display=True)
```

```python
h2.print_history()
```

```python
h2.plot_standard()
```

----
&nbsp;
#### Example 3

```python
# Hyperbola 3: Rectangular (coefficients of x and y are the same)
hyperbola3 = "(x-2)**2/4 - (y-3)**2/4 - 1"
h3 = Conic.create(hyperbola3)
h3.get_info()
```

```python
h3.translate()
```

```python
h3.print_history()
```

```python
h3.plot_standard()
```

----
&nbsp;
#### Example 4

```python
# Hyperbola 4: rotated by some angle theta
hyperbola4 = "3*(x-1)**2 + 6*(x-1)*(y+2) - (y+2)**2 - 5"
h4 = Conic.create(hyperbola4)
h4.get_info()
```

```python
h4.translate()
h4.rotate()
h4.print_history()
```

```python
h4.plot_standard(x_range=(-2, 2), y_range=(-2, 2))
```

----
&nbsp;
#### Example 5

```python
# Hyperbola 5: rotated by some angle theta
hyperbola5 = "5*(x+2)**2 - 4*(x+2)*(y-3) - 2*(y-1)**2 - 7"
h5 = Conic.create(hyperbola5)
h5.get_info()
```

```python
h5.translate()
h5.rotate()
h5.print_history()
```

```python
h5.plot_standard(x_range=(-2, 2), y_range=(-2, 2))
```

## We define a rectangular hyperbola and rotate by 60 degrees

```python
rect_1 = "x^2-y^2-1"
test = Conic.create(rect_1)
test.get_info()
```

```python
test.rotate(display=True, rotate_it=60)
test.plot()
```

```python
test.transverse_axis_line()
```

```python
test.print_history()
```

## Try a non-special angle, 50 degrees

```python
rect_2 = "x^2-y^2-1"
test2 = Conic.create(rect_2)
test2.get_info()
```

```python
test2.rotate(display=True, rotate_it=50)
test2.plot()
```

That's a very interesting general form equation!

```python
test2.print_history()
```

```python
rotation_angle = test2.orientation[1]
sympy.simplify(rotation_angle)
```
