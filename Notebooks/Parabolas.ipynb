## Rotated parabolas - presence of $xy$ term in general equation

To rotate a parabola parallel to coordinate axes through some angle $\theta$
&nbsp;
$$
(A\cos^2 \theta + C \sin^2 \theta)x^2 + (A-C)\sin(2\theta)\cdot xy + (A\cos^2 \theta + C \sin^2 \theta)y^2 + (D \cos \theta - E \sin \theta)x + (D \sin \theta + E \cos \theta)y + F
$$
so that
$$
\sqrt x + \sqrt y = \sqrt c
$$
&nbsp;
Where $\theta = \frac{\pi}{4}$ such that,
$$
x^2-2xy+y^2-2cx-2cy+c^2=0
$$

#### $\frac{\pi}{4}$ rotation systematically
&nbsp;
1. rotation from $x=y^2$
2. rotation from $y=x^2$
3. rotation from $x=-y^2$
4. rotation from $y=-x^2$

----
&nbsp;
The below equations have been multiplied out from above to eliminate radical coefficients,

```python
r1 = "x^2-2*x*y+y^2-4*x-4*y+4"
r2 = "x^2+2*x*y+y^2+4*x-4*y+4"
r3 = "x^2-2*x*y+y^2+4*x+4*y+4"
r4 = "x^2+2*x*y+y^2-4*x+4*y+4"
```

```python
rot1 = Conic.create(r1)
rot2 = Conic.create(r2)
rot3 = Conic.create(r3)
rot4 = Conic.create(r4)
```

```python
rot_parabolas = [rot1, rot2, rot3, rot4]
for parabola in rot_parabolas:
    parabola.get_info()
```

## Rotate the first instance

```python
rot1.rotate_parabola()
rot1.print_history()
rot1.plot()
```
