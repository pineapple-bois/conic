## Testing a few parabolas at once

```python
# FUNCTIONS OF X
# Parabola 1: General form equivalent to a parabola with vertex at (3, -2) and opens upwards
parabola1 = "((x - 3)**2) / 4 - 2"

# Parabola 2: General form equivalent to a parabola with vertex at (-5, 3) and opens downwards
parabola2 = "-((x + 5)**2) / 3 + 3"

# Parabola 3: General form equivalent to a parabola with vertex at (1, -4) and opens upwards
parabola3 = "((x - 1)**2) / 5 - 4"

# Parabola 4: Vertex at (3, 4) and opens downwards
parabola4 = "-(x - 3)**2 + 4"


# FUNCTIONS OF Y
# Parabola 5: Vertex at (2, 3) and opens rightwards
parabola5 = "(y - 3)**2 + 2"

# Parabola 6: Vertex at (1, -1) and opens leftwards
parabola6 = "-(y + 1)**2 - 1"
```

----
&nbsp;
### Instantiation

```python
p1 = Conic.create(parabola1)
p2 = Conic.create(parabola2)
p3 = Conic.create(parabola3)
p4 = Conic.create(parabola4)
p5 = Conic.create(parabola5)
p6 = Conic.create(parabola6)
```

```python
parabolas = [p1, p2, p3, p4, p5, p6]
for i, parabola in enumerate(parabolas, start=1):
    parabola.get_info()
```

## Rotated parabolas - presence of $xy$ term in general equation

Parabola rotated by $45 \degree$
&nbsp;
$\sqrt x + \sqrt y = \sqrt c$
&nbsp;
Such that,
$$
x^2-2xy+y^2-2cx-2cy+c^2=0
$$

## $45 \degree$ rotation systematically
&nbsp;
1. rotation from $x=y^2$
2. rotation from $y=x^2$
3. rotation from $x=-y^2$
4. rotation from $y=-x^2$

----
&nbsp;
Declare equations in general form,

```python
r1 = "x^2-2*x*y+y^2-4*x-4*y+4"
r2 = "x^2+2*x*y+y^2+4*x-4*y+4"
r3 = "x^2-2*x*y+y^2+4*x+4*y+4"
r4 = "x^2+2*x*y+y^2-4*x+4*y+4"
```

```python
rot1 = Conic.create(r1)
rot2 = Conic.create(r2)
rot3 = Conic.create(r3)
rot4 = Conic.create(r4)
```

```python
rot_parabolas = [rot1, rot2, rot3, rot4]
for i, parabola in enumerate(rot_parabolas, start=1):
    parabola.get_info()
```
