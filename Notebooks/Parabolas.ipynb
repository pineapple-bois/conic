# Parabolas

```python
import sympy
from conics.conic_factory import *
```

----
&nbsp;
## Parabola aligned parallel to coordinate axes. A function of $x$
#### 1.

```python
# Parabola 1: General form equivalent to a parabola with vertex at (-5, 3) and opens downwards
parabola1 = "-((x + 5)**2) / 3 + 3"
p1 = Conic.create(parabola1)
p1.get_info()
```

```python
# We rotate to get it in standard orientation
p1.rotate(display=True)
p1.plot()
```

```python
# We map the vertex to the origin
p1.translate(display=True)
p1.plot_standard()
```

```python
# History of transformations applied
p1.print_history()
```

## Parabola rotated through some angle $\theta$
#### 3.

```python
r3 = "x^2/4 + x*y/3 + y^2/9 -2*x + 3*y -1"
rot3 = Conic.create(r3)
rot3.get_info()
```

## Zoom in

```python
rot3.plot(x_range=(-1, 1), y_range=(-1, 1))
```

## Rotate

```python
rot3.rotate()
rot3.plot()
```

## Find vertex of the original parabola.
Needs to be performed before translation to standard position

```python
rot3.vertex
```

```python
rot3.get_original_vertex(to_float=True)
```

```python
rot3.get_original_vertex()
```

```python
rot3.translate()
rot3.plot_standard(x_range=(-7, 7), y_range=(-7, 7))
```

```python
rot3.print_history()
```

### We can also rotate a parabola through some angle $\theta$ in degrees.
Depending on whether this is a 'special angle', the plotting method could have a long run time


```python
ptest = Conic.create("x^2")
ptest.get_info()
```

```python
ptest.rotate(display=True, rotate_it=30)
ptest.plot()
```

#### Using the `rotate(rotate_it=degrees)` method, we write orientation history *after* the rotation.
Therefore, if we rotate it back to align with one of the coordinate axes, we should be able to find the vertex of the rotated parabola!

```python
ptest.orientation_history
```

```python
ptest.rotate(display=True)
ptest.plot()
```

#### We then apply the `get_original_vertex()` method

```python
ptest.get_original_vertex()
```

Above was a trivial example.

#### Then print the history of the object

```python
ptest.print_history()
```

#### Testing a non-special angle

```python
ptest2 = Conic.create("y^2")
ptest2.get_info()
```

```python
ptest2.rotate(display=True, rotate_it=50)
ptest2.plot()
```

We reverse the operation

```python
ptest2.rotate(display=True, rotate_it=-50)
ptest2.plot()
```

```python
ptest2.print_history()
```

### These operations result in pretty spicy coefficients although the parabola is in standard form!
