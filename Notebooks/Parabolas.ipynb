## Testing a few parabolas at once

```python
# FUNCTIONS OF X
# Parabola 1: General form equivalent to a parabola with vertex at (3, -2) and opens upwards
parabola1 = "((x - 3)**2) / 4 - 2"

# Parabola 2: General form equivalent to a parabola with vertex at (-5, 3) and opens downwards
parabola2 = "-((x + 5)**2) / 3 + 3"

# Parabola 3: General form equivalent to a parabola with vertex at (1, -4) and opens upwards
parabola3 = "((x - 1)**2) / 5 - 4"

# Parabola 4: Vertex at (3, 4) and opens downwards
parabola4 = "-(x - 3)**2 + 4"


# FUNCTIONS OF Y
# Parabola 5: Vertex at (2, 3) and opens rightwards
parabola5 = "(y - 3)**2 + 2"

# Parabola 6: Vertex at (1, -1) and opens leftwards
parabola6 = "-(y + 1)**2 - 1"
```

----
&nbsp;
### Instantiation

```python
p1 = Conic.create(parabola1)
p2 = Conic.create(parabola2)
p3 = Conic.create(parabola3)
p4 = Conic.create(parabola4)
p5 = Conic.create(parabola5)
p6 = Conic.create(parabola6)
```

```python
parabolas = [p1, p2, p3, p4, p5, p6]
for i, parabola in enumerate(parabolas, start=1):
    print("---")
    print(f"{parabola.__repr__()}\nType: {parabola.type}\nInstance: {type(parabola)}"
          f"\nCoefficients: {parabola.coeff}\nGeneral Form: {parabola}")
    parabola.display()
    print(f"\nMatrix of Coefficients:")
    parabola.print_matrix()
    print(f"\nCompleted Sq. Form: {parabola.complete_square()}")
    print(f"\nVertex: {parabola.compute_vertex()}")
    parabola.plot()
```
