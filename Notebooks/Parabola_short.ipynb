## Rotated - Gives some spicy rational numbers...

```python
r1 = "x^2/4 + x*y/3 + y^2/9 -2*x + 3*y -1"
rot1 = Conic.create(r1)
rot1.get_info()
```

Zoom in

```python
rot1.plot_parabola(x_range=(-1, 1), y_range=(-1, 1))
```

Rotate

```python
rot1.rotate_parabola(rational=True)
rot1.plot_parabola()
```

Translate

```python
rot1.translate_origin(rational=True)
rot1.plot_standard(x_range=(-7, 7), y_range=(-7, 7))
```

What did we just do?

```python
rot1.print_history()
```

## Rotated - Coefficients give nicer rational numbers this time!

```python
r2 = "16*x^2 + 24*x*y + 9*y^2 -130*x - 90*y"
rot2 = Conic.create(r2)
rot2.get_info()
```

Does this look like a parabola? No. However, it's been instantiated as such. The determinant of the quadratic matrix is zero such that $B^2=4AC$.
&nbsp;
Therefore, we rotate.

```python
rot2.rotate_parabola(rational=True)
rot2.plot_parabola()
```

The $x$ and $y$ range can be tweaked to best display the characteristics

```python
rot2.translate_origin(rational=True)
rot2.plot_standard(x_range=(-0.2, 0.2), y_range=(-0.2, 0.2))
```

What did we do to get it as above?

```python
rot2.print_history()
```
